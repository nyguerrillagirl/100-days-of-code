# Activity 1.01: Using NumPy to compute the *mean*, *median*, *variance*, and *standard deviation* of a dataset

```python
import numpy as np
```

```python
dataset = np.genfromtxt('../Datasets/normal_distribution.csv', delimiter=',')
```

```python
# Print a subset of the first two rows of the dataset
print(dataset[0:2,:])
```

```python
# The mean of the third row
np.mean(dataset[2])
```

```python
# Index the last element of an ndarray
dataset[-1,-1]
```

```python
# Get a submatrix of the first three elements of every row of the first three column
dataset[0:3, 0:3]
```

```python
np.median(dataset[-1,:])
```

```python
dataset[:, -3:]
```

```python
np.median(dataset, axis=1)
```

```python
np.var(dataset, axis=0)
```

```python
np.var(dataset[-2:,:2])
```

```python
np.std(dataset)
```
