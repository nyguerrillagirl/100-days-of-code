### Exercise 1.02: Indexing, Slicing, Splitting, and Iterating

```python
import numpy as np
```

```python
dataset = np.genfromtxt('../Datasets/normal_distribution_splittable.csv', delimiter=',')
```

```python
second_row = dataset[1]
np.mean(second_row)
```

```python
last_row = dataset[-1]
np.mean(last_row)
```

```python
np.mean(dataset)
```

```python
# access the first value of the first row and the last value of the second last row
first_val_first_row = dataset[0][0]
np.mean(first_val_first_row)
```

```python
first_val_first_row
```

```python
last_val_second_last_row = dataset[-2, -1]
np.mean(last_val_second_last_row)
```

```python
last_val_second_last_row
```

```python
"""
slicing an intersection of 4 elements (2x2) of the first two rows
and the first two columns
"""
subsection_2x2 = dataset[1:3, 1:3]
```

```python
np.mean(subsection_2x2)
```

```python
# we want to have every other element of the fifth row
every_other_elem = dataset[4, ::2]
np.mean(every_other_elem)
```

```python
# reverse the elements in a slice using negative numbers:
reversed_last_row = dataset[-1, ::-1]
np.mean(reversed_last_row)
```

```python
hor_splits = np.hsplit(dataset,(3))
hor_splits[0].shape
```

```python
# split the first third into two equal parts vertically
ver_splits = np.vsplit(hor_splits[0],(2))
ver_splits
```

```python
print("Dataset", dataset.shape)
print("Subset", ver_splits[0].shape)
```

```python
# iterate over the whole dataset
curr_index = 0
for x in np.nditer(dataset):
    print(x, curr_index)
    curr_index += 1
```

```python
for index, value in np.ndenumerate(dataset):
    print(index, value)
```

```python
# (5, 6) 104.60344836
dataset[5,6]
```
