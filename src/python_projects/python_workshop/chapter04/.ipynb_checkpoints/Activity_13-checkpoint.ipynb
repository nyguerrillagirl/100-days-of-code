```python
# Load the CSV file
import csv
lines = []
with open('titanic_train.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for line in csv_reader:
        lines.append(line)
```

```python
# Prepare a data object that stores all the passenger details 
data = lines[1:]
passengers = []
headers = lines[0]
```

```python
print(headers)
```

```python
# Create a dictionary object for each passenger and put the dictionary into passengers list
for d in data:
    p = {}
    for i in range(0, len(headers)):
        key = headers[i]
        value = d[i]
        p[key] = value
    passengers.append(p)
```

```python
print(passengers[0])
```

```python
survived = [p['Survived'] for p in passengers]
pclass = [p['Pclass'] for p in passengers]
age = [float(p['Age']) for p in passengers if p['Age'] != '']
gender_survived = [p['Sex'] for p in passengers if int(p['Survived']) == 1]
```

```python
print(survived)
```

```python
# Visualize the proportion of passengers that survived the incident (in a pie chart)
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from collections import Counter
labels = ['0', '1']
plt.pie(Counter(survived).values(), labels=Counter(survived).keys(), autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'yellow'])
plt.title('Survived', fontdict={'fontsize':20})        
plt.show()
```

```python
plt.title("surviving passengers count by gender")
plt.bar(Counter(gender_survived).keys(), Counter(gender_survived).values())
plt.show()
```
