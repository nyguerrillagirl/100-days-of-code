# Activity 25: Using Machine Learning to Predict Customer Return Rate Accuracy

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('CHURN.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
# Replace 'Yes', 'No' in Churn column
df['Churn'] = df['Churn'].replace(to_replace=['No', 'Yes'], value=[0, 1])
```

```python
df.head()
```

```python
# all cols except first and last
X = df.iloc[:, 1:-1]
# last col only
y = df.iloc[:, -1]
```

```python
X.head()
```

```python
y.head()
```

```python
# Transform all the predictive columns into numeric columns
X = pd.get_dummies(X)
```

```python
X.head()
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
```

```python
# Write a classifier function that takes a model as its input
def clf_model(model):
    clf = model
    scores = cross_val_score(clf, X, y)
    print('Scores:', scores)
    print('Mean score:', scores.mean())
```

```python
# Run the clf_model function with LogisticRegression() as input
clf_model(LogisticRegression(max_iter=1000))
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf_model(KNeighborsClassifier())
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf_model(KNeighborsClassifier())
```

```python
from sklearn.naive_bayes import GaussianNB
clf_model(GaussianNB())
```

```python
from sklearn.ensemble import RandomForestClassifier
clf_model(RandomForestClassifier())
```

```python
from sklearn.ensemble import AdaBoostClassifier
clf_model(AdaBoostClassifier())
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test ,y_train, y_test = train_test_split(X, y, test_size = 0.25)
def confusion(model):
    clf = model
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print('Confusion Matrix:', confusion_matrix(y_test, y_pred))
    print('Classfication Report:', classification_report(y_test, y_pred))
    
    return clf
```

```python
confusion(AdaBoostClassifier())
```

```python
confusion(RandomForestClassifier())
```

```python
confusion(LogisticRegression())
```

```python
confusion(AdaBoostClassifier(n_estimators=250))
```

```python
confusion(AdaBoostClassifier(n_estimators=25))
```

```python
confusion(AdaBoostClassifier(n_estimators=15))
```
