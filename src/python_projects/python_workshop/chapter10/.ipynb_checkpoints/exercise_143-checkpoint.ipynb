### Exercise 143: Correlation Values from the Dataset

```python
import pandas as pd
housing_df = pd.read_csv('HousingData.csv')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
housing_df.corr()
```

```python
import seaborn as sns
# Set up seaborn dark grid
sns.set()
```

```python
corr = housing_df.corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, xticklabels=corr.columns.values, yticklabels=corr.columns.values, cmap="Blues", linewidths=1.25, alpha=0.8)
plt.show()
```

```python
x = housing_df['RM']
x = pd.DataFrame(x)
y = housing_df['MEDV']
y = pd.DataFrame(y)
df_x_y = pd.concat([x,y])
df_x_y
df_x_y = df_x_y.fillna(df_x_y.median())
```

```python
df = pd.DataFrame({'x': [1, 2, 3, 4, 5], 'y': [2, 4, 5, 4, 6]})
plt.figure(figsize=(10, 7))
sns.regplot(data=df, x='x', y='y')
plt.show()
```
